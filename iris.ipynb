```python
import pandas as pd
df=pd.read_csv("iris.csv")
df
```

```python
print(df.dtypes)
```

```python
df['species']=df['species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
# Set the style for the plot
sns.set(style="whitegrid")
# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='viridis', s=80)

# Set plot labels and title
plt.title('Scatter Plot of Sepal Length vs Sepal Width', fontsize=16)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)
# Show the legend
plt.legend(title='Species', loc='upper right')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
# Set the style for the plot
sns.set(style="whitegrid")
# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, palette='viridis', s=80)

# Set plot labels and title
plt.title('Scatter Plot of Sepal Length vs Sepal Width', fontsize=16)
plt.xlabel('petal Length', fontsize=12)
plt.ylabel('petalal Width', fontsize=12)
# Show the legend
plt.legend(title='Species', loc='upper right')

# Show the plot
plt.show()
```

```python
X=df.drop('species',axis=1)
print(X)
Y=df['species']
print(Y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
Xt,Xtest,Yt,Ytest = train_test_split(X,Y, test_size=0.2)#spliting the data
model = LogisticRegression()#creating the model
model.fit(Xt,Yt)#training the model
predictions = model.predict(Xt)#predicting
print("Y_test shape:", Yt.shape)
print("Predictions shape:", predictions.shape)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Yt,predictions)#finding the accuracy
print(accuracy)
```
